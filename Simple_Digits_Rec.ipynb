## SIMPLE DEEP LEARNING PROJECT: Handwritten Digit Recognition
### Learn to recognize handwritten digits (0-9) using a neural network!

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
print("=" * 60)
print("SIMPLE DEEP LEARNING: Handwritten Digit Recognition")
print("=" * 60)
```

```python
# STEP 1: Load the data
print("\nüìö STEP 1: Loading handwritten digit images...")
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print(f"   ‚úì Loaded {len(x_train)} training images")
print(f"   ‚úì Loaded {len(x_test)} test images")
print(f"   ‚úì Each image is 28x28 pixels")
```

```python
# Show some example images
print("\nüì∏ Let's see some example digits:")
plt.figure(figsize=(10, 2))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(str(y_train[i]))
    plt.axis('off')
plt.suptitle('Sample Handwritten Digits', fontsize=14)
plt.tight_layout()
plt.savefig('sample_digits.png', dpi=150, bbox_inches='tight')
print("   ‚úì Sample images saved as 'sample_digits.png'")
```

```python
# STEP 2: Prepare the data
print("\nüîß STEP 2: Preparing the data...")
```

```python
# Flatten images from 28x28 to 784 (one long array)
x_train = x_train.reshape(-1, 784).astype('float32') / 255.0
x_test = x_test.reshape(-1, 784).astype('float32') / 255.0
```

```python
print("   ‚úì Reshaped images to 784 pixels")
print("   ‚úì Normalized pixel values (0 to 1)")
```

```python
# STEP 3: Build the neural network
print("\nüß† STEP 3: Building the neural network...")
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(784,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

print("\n   Neural Network Architecture:")
print("   " + "-" * 40)
print("   Input Layer:    784 neurons (28x28 pixels)")
print("   Hidden Layer 1: 128 neurons (with ReLU)")
print("   Hidden Layer 2: 64 neurons (with ReLU)")
print("   Output Layer:   10 neurons (digits 0-9)")
print("   " + "-" * 40)
```

```python
# STEP 4: Compile the model
print("\n‚öôÔ∏è  STEP 4: Configuring the model...")
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
print("   ‚úì Optimizer: Adam (smart learning algorithm)")
print("   ‚úì Loss: Cross-entropy (for classification)")
print("   ‚úì Metric: Accuracy")
```

```python
# STEP 5: Train the model
print("\nüèãÔ∏è  STEP 5: Training the model...")
print("   This will take about 1-2 minutes...\n")

history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

print("\n   ‚úì Training complete!")
```

```python
# STEP 6: Evaluate the model
print("\nüìä STEP 6: Testing the model...")
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)

print(f"\n   {'='*40}")
print(f"   FINAL RESULTS:")
print(f"   {'='*40}")
print(f"   Test Accuracy: {test_accuracy*100:.2f}%")
print(f"   Test Loss:     {test_loss:.4f}")
print(f"   {'='*40}")

if test_accuracy > 0.95:
    print("   üéâ Excellent! Your model is very accurate!")
elif test_accuracy > 0.90:
    print("   üëç Good job! Your model works well!")
else:
    print("   üìà Not bad for a first try!")
```

```python
# STEP 7: Visualize training progress
print("\nüìà STEP 7: Creating training charts...")
plt.figure(figsize=(12, 4))
```

```python
# Accuracy chart
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training', linewidth=2, color='blue')
plt.plot(history.history['val_accuracy'], label='Validation', linewidth=2, color='orange')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Model Accuracy Over Time', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
# Loss chart
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training', linewidth=2, color='blue')
plt.plot(history.history['val_loss'], label='Validation', linewidth=2, color='orange')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.title('Model Loss Over Time', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
plt.tight_layout()
plt.savefig('training_results.png', dpi=150, bbox_inches='tight')
print("   ‚úì Charts saved as 'training_results.png'")
```

```python
# STEP 8: Make predictions
print("\nüîÆ STEP 8: Making predictions on new images...")
```

```python
# Get 20 random test images
indices = np.random.choice(len(x_test), 20, replace=False)
predictions = model.predict(x_test[indices], verbose=0)
```

```python
# Visualize predictions
plt.figure(figsize=(15, 6))
for i in range(20):
    plt.subplot(4, 5, i+1)
    
    # Reshape back to 28x28 for display
    image = x_test[indices[i]].reshape(28, 28)
    plt.imshow(image, cmap='gray')
    
    # Get predicted and true labels
    predicted = np.argmax(predictions[i])
    actual = y_test[indices[i]]
    
    # Color: green if correct, red if wrong
    color = 'green' if predicted == actual else 'red'
    
    plt.title(f'Predicted: {predicted}\nActual: {actual}', 
              color=color, fontsize=9)
    plt.axis('off')

plt.suptitle('Model Predictions (Green = Correct, Red = Wrong)', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('predictions.png', dpi=150, bbox_inches='tight')
print("   ‚úì Predictions saved as 'predictions.png'")
```

```python
# Calculate accuracy on these samples
correct = sum([np.argmax(predictions[i]) == y_test[indices[i]] for i in range(20)])
print(f"   ‚úì Got {correct} out of 20 correct ({correct*5}%)")
```
